```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
import tensorflow as tf

seed=0
np.random.seed(seed)
tf.random.set_seed(seed)

IN_data = np.loadtxt("C:\work\data-03-diabetes.csv", delimiter=",")
x_data = IN_data[:, 0:-1]
y_data = IN_data[:, -1]

model = Sequential()
model.add(Dense(30, input_dim=8, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(1, activation='sigmoid'))               # 0 아니면 1이므로 sigmoid 사용

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_data, y_data, epochs=30, batch_size=10)

print(model.evaluate(x_data, y_data))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

np.random.seed(5)

dataset = np.loadtxt("C:\work\data-03-diabetes.csv", delimiter=",", dtype=np.float32)

x_train = dataset[:700, 0:8]
y_train = dataset[:700, 8]

x_test = dataset[700:, 0:8]
y_test = dataset[700, 8]

model = Sequential()
model.add(Dense(17, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))               # 0 아니면 1이므로 sigmoid 사용

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=1500, batch_size=64)

# print(model.evaluate(x_data, y_data))
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
